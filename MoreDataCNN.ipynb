```python
import torch
import torchvision.transforms.functional as TF
from torchvision.transforms.functional import InterpolationMode

def resize_tensor(tensor, target_size=(12, 12), mode='bilinear'):
    """
    tensor:      (H, W, C)   OR (C, H, W)
    target_size: (H_out, W_out)
    mode:        'nearest' | 'bilinear' | 'bicubic'
    returns:     (C, H_out, W_out)
    """
    # If tensor is HWC, move channels to the front
    if tensor.ndim == 3 and tensor.shape[0] != tensor.shape[-1]:
        # heuristic: assume HWC when first dim smaller than last
        if tensor.shape[0] < tensor.shape[-1]:
            tensor = tensor.permute(2, 0, 1)     # HWC -> CHW

    interp = {
        'nearest':  InterpolationMode.NEAREST,
        'bilinear': InterpolationMode.BILINEAR,
        'bicubic':  InterpolationMode.BICUBIC
    }[mode]

    tensor = TF.resize(tensor, target_size,
                       interpolation=interp,
                       antialias=True)

    return tensor        # CHW

```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch

def show_tensor(tensor, title=None, channels=None):
    """
    Display selected channels of a tensor using matplotlib.
    
    Accepts:
        • (C, H, W)  : channel-first format
        • (H, W, C)  : channel-last format
        • (H, W)     : single-channel image
    """
    # Move to CPU and ensure it's no longer a gradient tensor
    tensor = tensor.detach().cpu()
    
    # Check tensor shape and dimensions
    if tensor.ndim == 3:
        if tensor.shape[0] <= tensor.shape[-1]:  # HWC
            img = tensor.numpy()  # Already in HWC
        else:  # CHW
            img = tensor.permute(1, 2, 0).numpy()  # Convert to HWC
        
        # If specific channels are provided, select them
        if channels is not None:
            img = img[:, :, channels]  # Select the specific channels
    elif tensor.ndim == 2:
        img = tensor.numpy()  # Treat as HWC with single channel
    else:
        raise ValueError("Tensor must be 2D or 3D, got shape {}".format(tensor.shape))

    # Prepare the figure for channels to visualize
    if img.ndim == 3:
        num_channels = img.shape[-1]
    else:
        num_channels = 1
    
    # Create subplots for each channel
    fig, axs = plt.subplots(1, num_channels, figsize=(5 * num_channels, 5))
    
    if num_channels == 1:
        axs = [axs]  # Make it iterable
    
    for i in range(num_channels):
        channel_img = img[:, :, i] if num_channels > 1 else img
        
        # Normalize the channel to [0, 1] range for visualization
        img_min = channel_img.min()
        img_max = channel_img.max()
        if img_max > img_min:  # Avoid divide-by-zero on constant images
            channel_img = (channel_img - img_min) / (img_max - img_min)
        
        # If it's a single channel, just use a grayscale colormap
        axs[i].imshow(channel_img, cmap='jet')
        axs[i].axis('off')
        axs[i].set_title(f"Channel {i+1}")

    plt.suptitle(title)
    plt.show()

# Example Usage
shape_id = 'tree_crown_1'  # replace as needed
tensor_to_show = resized_crowns[shape_id]  # CHW (30, 512, 512)

# Visualize selected channels: replace with the indices of channels you want
show_tensor(tensor_to_show, title=shape_id, channels=[4])  # Assuming channels 0, 1, 2 are what you want
```

```python
import pandas as pd
import torch

# Assuming resized_crowns is your dictionary with tensors
# Example: resized_crowns = {'tree_crown_1': torch.randn(27, 15, 15), ...}

# Load your CSV data into a pandas DataFrame
# df = pd.read_csv('your_file.csv')  # Example loading of DataFrame

# Create a mapping from shapeID to pft
shapeid_to_pft = df[['shapeID', 'pft']].drop_duplicates().set_index('shapeID')['pft'].to_dict()

# Step 1: Attach labels to tensors while filtering for channel count
data_list = []
labels_list = []

for shapeID, tensor in resized_crowns.items():
    print(f"Processing {shapeID} with shape {tensor.shape}")  # Debugging line
    # Check if the tensor has 27 channels
    if tensor.shape[0] == 26:  # Check the number of channels
        # Retrieve the pft label for the current shapeID
        pft_label = shapeid_to_pft.get(shapeID)

        if pft_label is not None:
            data_list.append(tensor.unsqueeze(0))  # Add a new dimension for batching
            labels_list.append(pft_label)
        else:
            print(f"No pft label found for {shapeID}")  # Debugging line
    else:
        print(f"Skipping {shapeID} due to incorrect channel count")  # Debugging line

# Check if data_list is empty
if not data_list:
    print("No valid tensors found. Please check the conditions and data.")

# Now the data_list and labels_list will only contain tensors with 27 channels
print(f"Number of valid tensors: {len(data_list)}")
print(f"Number of valid labels: {len(labels_list)}")

# Convert lists to tensors if they are not empty
if data_list:
    X = torch.cat(data_list, dim=0)  # Shape: (N, 27, 15, 15) where N is the number of samples

    # Convert string labels to a list of integers
    unique_pfts = {pft: idx for idx, pft in enumerate(set(labels_list))}  # Create a mapping from pft to integer
    y = torch.tensor([unique_pfts[label] for label in labels_list], dtype=torch.int64)  # Convert to tensor of integers

    # Final shapes
    print(f"Input tensor shape: {X.shape}")  # Should be (N, 27, 15, 15)
    print(f"Labels tensor shape: {y.shape}")  # Should be (N,)
    print(f"Unique PFT classes: {unique_pfts}")
else:
    print("No data to process.")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class CustomTensorDataset(Dataset):
    def __init__(self, tensors, labels, transform=None):
        self.tensors = tensors
        self.labels = labels
        self.transform = transform
        
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        tensor = self.tensors[idx]
        label = self.labels[idx]

        # Apply transformations if specified
        if self.transform:
            tensor = self.transform(tensor)
        
        return tensor, label

# Define your custom transformations
class CustomTransform:
    def __call__(self, img_tensor):
        # Example of simple random horizontal flip
        if torch.rand(1).item() > 0.5:
            img_tensor = torch.flip(img_tensor, [2])  # Flips horizontally
        # You can add more custom transformations here (rotation, normalization, etc.)
        return img_tensor

# Shuffle the dataset before splitting
indices = torch.randperm(len(X))  # Assuming X is your input tensor of shape [num_samples, 30, H, W]
X_shuffled = X[indices]           # Shuffle features
y_shuffled = y[indices]           # Shuffle labels

# Create datasets
train_size = int(0.8 * len(X))
train_data = CustomTensorDataset(X_shuffled[:train_size], y_shuffled[:train_size], transform=CustomTransform())
val_data = CustomTensorDataset(X_shuffled[train_size:], y_shuffled[train_size:])

# DataLoaders
train_dataloader = DataLoader(
    train_data,
    batch_size=10,
    shuffle=True  
)

validation_dataloader = DataLoader(
    val_data,
    batch_size=10,
    shuffle=False  
)

# Your training loop remains the same
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self, num_classes):
        super(SimpleCNN, self).__init__()
        
        # Define convolutional layers
        self.conv1 = nn.Conv2d(in_channels=26, out_channels=16, kernel_size=3, padding=1)  # Output: (16, 15, 15)
        self.bn1 = nn.BatchNorm2d(16)  # Batch normalization after conv1

        self.conv2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3, padding=1)  # Output: (32, 15, 15)
        self.bn2 = nn.BatchNorm2d(32)  # Batch normalization after conv2

        self.conv3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3, padding=1)  # Output: (64, 15, 15)
        self.bn3 = nn.BatchNorm2d(64)  # Batch normalization after conv3

        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)  # Pooling to reduce dimensions
        
        # Define the fully connected layers
        self.fc1 = nn.Linear(64 * 1 * 1, 128)  # Adjusted input size based on previous dimensions
        self.dropout1 = nn.Dropout(0.5)  # Dropout layer after the first fully connected layer
        self.fc2 = nn.Linear(128, num_classes)  # Output layer
        self.dropout2 = nn.Dropout(0.5)  # Optional dropout before the final output layer

    def forward(self, x):
        # Forward pass through the model
        x = self.pool(F.relu(self.bn1(self.conv1(x))))  # Output: (16, 7, 7)
        x = self.pool(F.relu(self.bn2(self.conv2(x))))  # Output: (32, 3, 3)
        x = self.pool(F.relu(self.bn3(self.conv3(x))))  # Output: (64, 1, 1)
        
        # Flatten for the fully connected layer
        x = x.view(-1, 64 * 1 * 1)  # Flatten to (batch_size, 64)
        
        # Fully connected layers with dropout
        x = self.dropout1(F.relu(self.fc1(x)))
        x = self.dropout2(self.fc2(x))  # No activation on output layer if using cross-entropy loss.
        
        return x

# Instantiate the model
num_classes = len(unique_pfts)  # Number of unique classes (PFTs)
model = SimpleCNN(num_classes)
```

```python
### save best model:
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader

# Assuming the model, criterion, optimizer, train_dataloader, validation_dataloader, and num_epochs are defined.
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)

train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Step 3: Training Loop
num_epochs = 100  # Set the number of epochs as needed
best_val_accuracy = 0.0  # Initialize best validation accuracy

for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0
    correct = 0
    total = 0
    
    for batch_tensors, batch_labels in train_dataloader:
        optimizer.zero_grad()  # Zero the gradients
        outputs = model(batch_tensors)  # Forward pass
        loss = criterion(outputs, batch_labels)  # Calculate loss
        loss.backward()  # Backward pass
        optimizer.step()  # Update weights
        
        running_loss += loss.item()
        
        # Calculate accuracy for the batch
        _, predicted = torch.max(outputs.data, 1)
        total += batch_labels.size(0)  # Total instances in this batch
        correct += (predicted == batch_labels).sum().item()  # Count correct predictions
    
    average_loss = running_loss / len(train_dataloader)  # Average loss for the epoch
    average_accuracy = 100 * correct / total  # Training accuracy percentage
    train_losses.append(average_loss)
    train_accuracies.append(average_accuracy)

    # Validation Routine
    model.eval()  # Set model to evaluation mode
    val_running_loss = 0.0
    val_correct = 0
    val_total = 0
    
    with torch.no_grad():  # Disable gradient calculation
        for val_batch_tensors, val_batch_labels in validation_dataloader:
            val_outputs = model(val_batch_tensors)  # Forward pass for validation
            val_loss = criterion(val_outputs, val_batch_labels)  # Calculate validation loss
            val_running_loss += val_loss.item()

            # Calculate accuracy for the validation batch
            _, val_predicted = torch.max(val_outputs.data, 1)
            val_total += val_batch_labels.size(0)
            val_correct += (val_predicted == val_batch_labels).sum().item()

    val_average_loss = val_running_loss / len(validation_dataloader)  # Average validation loss for this epoch
    val_average_accuracy = 100 * val_correct / val_total  # Validation accuracy percentage
    val_losses.append(val_average_loss)
    val_accuracies.append(val_average_accuracy)

    print(f'Epoch [{epoch + 1}/{num_epochs}], '
          f'Train Loss: {average_loss:.4f}, Train Accuracy: {average_accuracy:.2f}%, '
          f'Val Loss: {val_average_loss:.4f}, Val Accuracy: {val_average_accuracy:.2f}%')

    # Save the model if the validation accuracy is the best we've seen so far
    if val_average_accuracy > best_val_accuracy:
        best_val_accuracy = val_average_accuracy
        torch.save(model.state_dict(), 'best_model.pth')
        print(f'New best model saved with validation accuracy: {best_val_accuracy:.2f}%')

# After the training loop, you can plot the training history as discussed before
```

```python
import matplotlib.pyplot as plt

def plot_training_history(train_losses, val_losses, train_accuracies, val_accuracies):
    epochs = range(1, len(train_losses) + 1)

    # Plot loss
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, label='Training Loss')
    plt.plot(epochs, val_losses, label='Validation Loss')
    plt.title('Loss over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accuracies, label='Training Accuracy')
    plt.plot(epochs, val_accuracies, label='Validation Accuracy')
    plt.title('Accuracy over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

def visualize_filters(layer):
    # Assume layer is a Conv2d layer
    filters = layer.weight.data.clone()  # Get the weights of the layer
    
    # Normalize filters for visualization
    filters = (filters - filters.min()) / (filters.max() - filters.min())
    
    num_filters = filters.shape[0]  # Number of filters
    num_channels = filters.shape[1]  # Number of input channels
    fig, axes = plt.subplots(num_filters, num_channels, figsize=(num_channels * 2, num_filters * 2))
    
    for i in range(num_filters):
        for j in range(num_channels):
            # Select the i-th filter and j-th channel
            axes[i, j].imshow(filters[i, j].cpu().numpy(), cmap='gray')  # Plot each channel
            axes[i, j].axis('off')
    
    plt.show()

# Example usage
visualize_filters(model.conv1)  # Visualize filters from the first convolutional layer
```

```python
num_features = 26
grid_size = 1000

# Convert the DataFrame to a NumPy array and reshape
data_array = df_test.to_numpy().reshape((grid_size, grid_size, num_features))

# Step 2: Prepare for patch extraction
patch_size = 15
half_patch = patch_size // 2

# Pad the data to handle edges
padded_data = np.pad(data_array, ((half_patch, half_patch), (half_patch, half_patch), (0, 0)), mode='reflect')
```

```python
patches = []

# Step 3: Extract patches
for i in range(grid_size):
    for j in range(grid_size):
        # Extract a 15x15 patch centered at (i, j)
        patch = padded_data[i:i + patch_size, j:j + patch_size, :]
        
        # Convert patch to a PyTorch tensor and add batch dimension
        patch_tensor = torch.tensor(patch, dtype=torch.float32).unsqueeze(0).permute(0, 3, 1, 2)
        
        # Store the patch tensor
        patches.append(patch_tensor)
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Assuming 'model' is already defined and loaded with weights
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)  # Move model to the appropriate device
model.eval()  # Set the model to evaluation mode

# Concatenate all patches into a single tensor
all_patches_tensor = torch.cat(patches, dim=0)  # Shape: (N, 26, 15, 15)

# Create a DataLoader for batch processing
batch_size = 32  # Adjust as needed
dataset = TensorDataset(all_patches_tensor)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

outputs = []
total_batches = len(dataloader)
progress_interval = max(1, total_batches // 10)  # Adjust progress update frequency

# Disable gradient computation for inference
with torch.no_grad():
    for batch_idx, batch in enumerate(dataloader):
        # Move batch to the appropriate device
        batch = batch[0].to(device)  # batch is a tuple, so we access the tensor with [0]
        
        # Forward pass to get predictions
        predictions = model(batch)
        
        # If you need class predictions, use torch.max
        _, predicted_classes = torch.max(predictions, 1)
        
        # Move predictions to CPU if needed for further processing
        outputs.append(predicted_classes.cpu())
        
        # Print progress at regular intervals
        if (batch_idx + 1) % progress_interval == 0 or (batch_idx + 1) == total_batches:
            print(f"Processed {batch_idx + 1}/{total_batches} batches ({(batch_idx + 1) / total_batches * 100:.2f}%)")

# Process outputs as needed
# Example: Convert outputs to a single tensor or list
outputs_tensor = torch.cat(outputs, dim=0)

# If needed, reshape outputs_tensor to match the original grid size
# Example: outputs_tensor.view(grid_size, grid_size)
```
